## Part 2: Explore one Wikipedia page with the beautifulsoup package

```python
%pip install requests pandas

import bs4
import requests
import pandas as pd
```

```python
#save and print the text content of a page with all tags removed
```

```python
# URL of the Wikipedia page
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)'

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = bs4.BeautifulSoup(response.content, 'html.parser')

# Extract the text content from the parsed HTML
text_content = soup.get_text()

# Save the text content to a file
with open('list_of_countries_by_gdp_nominal.txt', 'w', encoding='utf-8') as file:
    file.write(text_content)

# Print the text content
print(text_content)
```

```python
#download an image with beautifulsoup and save it in this repository
```

```python
from urllib.parse import urljoin

# URL of the Wikipedia page
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)'

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = bs4.BeautifulSoup(response.content, 'html.parser')

# Find the first image in the page
image = soup.find('img')

# Get the URL of the image
image_url = urljoin(url, image['src'])

# Send a GET request to the image URL
image_response = requests.get(image_url)

# Save the image to a file
with open('gdp_nominal_image.jpg', 'wb') as file:
    file.write(image_response.content)

print("Image downloaded and saved as 'gdp_nominal_image.jpg'")
```

```python
# URL of the Wikipedia page
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)'

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = bs4.BeautifulSoup(response.content, 'html.parser')

# Find all the links in the page
links = soup.find_all('a', href=True)

# Print the first 100 characters of ten of these links
for link in links[:10]:
    print(link['href'][:100])
```

## Part 3: Downloading scripts

```python
scripts=pd.read_csv('pudding_data.csv')
```

```python
#using the links in the "link" column, download the first 1000 characters of each script
#use requests and bs4, remember to remove all html tags
```

```python
# Initialize a list to store the text of each script
script_texts = []

# Iterate over the links in the "link" column
for link in scripts['link']:
    # Send a GET request to the link
    response = requests.get(link)
    
    # Parse the HTML content using BeautifulSoup
    soup = bs4.BeautifulSoup(response.content, 'html.parser')
    
    # Extract the text content from the parsed HTML
    text_content = soup.get_text()
    
    # Get the first 1000 characters of the text content
    script_text = text_content[:1000]
    
    # Append the script text to the list
    script_texts.append(script_text)

# Add a new column to the dataframe with the text downloaded
scripts['script_text'] = script_texts

# Save the new dataframe as "pudding_texts.csv"
scripts.to_csv('pudding_texts.csv', index=False)
```

## Part 4: TMDB database

#### Browse the documentation at https://developer.themoviedb.org/reference/intro/getting-started. Create an account to authenticate

```python
#create a dataset of the movies in theaters now. Include metadata fields you are interested in. 
```

```python
import requests
import pandas as pd

# Replace 'YOUR_API_KEY' with your actual TMDB API key
api_key = 'YOUR_API_KEY'

# URL for the TMDB API endpoint to get movies in theaters
url = f'https://api.themoviedb.org/3/movie/now_playing?api_key={api_key}&language=en-US&page=1'

# Send a GET request to the URL
response = requests.get(url)
data = response.json()

# Extract the list of movies
movies = data['results']

# Create a dataframe with the metadata fields you are interested in
df_movies = pd.DataFrame(movies, columns=['title', 'release_date', 'overview', 'vote_average', 'poster_path'])

# Save the dataframe to a CSV file
df_movies.to_csv('movies_in_theaters.csv', index=False)

print(df_movies)
```

```python
#download the movie posters for 10 of these movies and save them to this repository
```

```python
# Base URL for the poster images
base_url = 'https://image.tmdb.org/t/p/w500'

# Download the posters for the first 10 movies
for index, row in df_movies.head(10).iterrows():
    poster_url = base_url + row['poster_path']
    poster_response = requests.get(poster_url)
    
    # Save the poster image to a file
    with open(f"poster_{index}.jpg", 'wb') as file:
        file.write(poster_response.content)

print("Posters downloaded and saved.")
```
